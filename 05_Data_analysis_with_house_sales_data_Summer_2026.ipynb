### Question 8: Feature Selection based on Correlation (2 Marks)

Calculate the correlation of all numerical features with the 'Price' column. Create a list named `highly_correlated_features` containing the names of features that have a correlation with 'Price' greater than 0.5. Print this list.

**Hint:** First calculate the full correlation matrix with `.corr()`. Then, select the 'Price' column from it. Filter this series to find features where the correlation value is > 0.5 and get their index.

**Expected Output:** A list of strings, where each string is the name of a highly correlated feature.

```python
# Write your answer here
```

```python
correlation_matrix = house_data.corr(numeric_only=True)
price_correlation = correlation_matrix['Price'].sort_values(ascending=False)

highly_correlated_features = price_correlation[price_correlation > 0.5].index.tolist()

# Exclude 'Price' itself from the feature list
if 'Price' in highly_correlated_features:
    highly_correlated_features.remove('Price')

print(highly_correlated_features)
```

### Question 9: Prepare Data for Modeling (1 Marks)

Using the `highly_correlated_features` list from the previous question, create your feature matrix `X`. Your target vector `y` should be the 'Price' column. Then, split `X` and `y` into training and testing sets with a `test_size` of 0.3 and a `random_state` of 123.

**Expected Output:** No direct output. The variables `X_train`, `X_test`, `y_train`, and `y_test` will be created.

```python
# Write your answer here
```

```python
from sklearn.model_selection import train_test_split

X = house_data[highly_correlated_features]
y = house_data['Price']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=123)

print(f"X_train shape: {X_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"y_test shape: {y_test.shape}")
```

### Question 10: Train and Evaluate a Linear Regression Model (5 Marks)

Import `LinearRegression` from `sklearn.linear_model` and `r2_score` from `sklearn.metrics`. Train a linear regression model on your training data. Make predictions on the test data and print the R-squared score.

**Expected Output:** A single decimal number representing the R-squared score of the model.

```python
# Write your answer here
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

model = LinearRegression()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)
r_squared = r2_score(y_test, y_pred)

print(f"{r_squared:,.3f}")
```

### Question 11: K-Means Clustering (5 Marks)

Import `KMeans` from `sklearn.cluster`. Initialize and fit the K-Means clustering model on the feature matrix `X` (from Question 9). Choose a suitable number of clusters (e.g., 4) to group similar houses. Assign the cluster labels to a variable `kmeans_labels`.

**Expected Output:** A successfully fitted K-Means model and the cluster labels array.

```python
from sklearn.cluster import KMeans

# Initialize and fit K-Means with 4 clusters (as an example for house data)
kmeans = KMeans(n_clusters=4, random_state=42, n_init='auto')
kmeans.fit(X)

kmeans_labels = kmeans.labels_

print(f"K-Means labels created with {len(np.unique(kmeans_labels))} unique clusters.")
```

### Question 12: F1 Score Evaluation for Clustering (5 Marks)

To evaluate the clustering, we will create 'pseudo-true' labels by categorizing the `Price` (`y`) into 4 equal-sized bins to match the number of clusters. Then, map each cluster label to the most frequent 'pseudo-true' label within that cluster. Finally, calculate and print the macro-averaged F1 score using `sklearn.metrics.f1_score`.

**Expected Output:** The calculated F1 score of the clustering.

```python
from sklearn.metrics import f1_score
import numpy as np

# Create 'pseudo-true' labels by binning the 'Price' (y) into 4 equal-sized bins
bins = pd.qcut(y, q=4, labels=False, duplicates='drop')

# Map cluster labels to pseudo-true labels
# Create a mapping dictionary: cluster_label -> most_frequent_bin
mapping = {}
for i in np.unique(kmeans_labels):
    mask = (kmeans_labels == i)
    # Find the most frequent 'pseudo-true' label within this cluster
    if not bins[mask].empty:
        most_frequent_bin = bins[mask].mode()[0]
        mapping[i] = most_frequent_bin

# Apply the mapping to the kmeans_labels to get aligned_kmeans_labels
aligned_kmeans_labels = pd.Series(kmeans_labels).map(mapping).values

# Calculate the F1 score. Use 'macro' average as the labels are not necessarily balanced.
f1 = f1_score(bins, aligned_kmeans_labels, average='macro')

print(f"Macro-averaged F1 Score: {f1:.3f}")
```

### Question 13: Confusion Matrix for Clustering (3 Marks)

Generate and display the confusion matrix comparing your mapped cluster labels (from Question 12) to the 'pseudo-true' binned price labels. Use `confusion_matrix` from `sklearn.metrics`. This will show how well your clusters align with the price categories.

**Expected Output:** A 4x4 confusion matrix printed as an array or a Pandas DataFrame.

```python
from sklearn.metrics import confusion_matrix
import pandas as pd

# Generate the confusion matrix
cm = confusion_matrix(bins, aligned_kmeans_labels)

# For better readability, put it in a DataFrame
cm_df = pd.DataFrame(cm, index=[f'True Bin {i}' for i in range(4)], columns=[f'Predicted Cluster {i}' for i in range(4)])

print("Confusion Matrix:")
print(cm_df)
```
